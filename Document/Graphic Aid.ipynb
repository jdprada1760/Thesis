# Triangles and rotations

```python
def genTriangle():
    a = [1-2*np.random.random(),1-2*np.random.random()]
    b = [1-2*np.random.random(),1-2*np.random.random()]
    c = [1-2*np.random.random(),1-2*np.random.random()]
    return np.array([a,b,c])

def drawTriangle(tr, color):
    plt.plot([tr[0,0],tr[1,0]],[tr[0,1],tr[1,1]], c= color)
    plt.plot([tr[1,0],tr[2,0]],[tr[1,1],tr[2,1]], c= color)
    plt.plot([tr[0,0],tr[2,0]],[tr[0,1],tr[2,1]], c= color)
    cm = (tr[0]+tr[1]+tr[2])/3.
    plt.scatter(0,0, c = 'b', s = 5)
    plt.scatter(cm[0],cm[1], c = color, s= 50, alpha = 0.5)
    lim = 0.3
    plt.xlim(min(tr.T[0])-lim,max(tr.T[0])+lim)
    plt.ylim(min(tr.T[1])-lim,max(tr.T[1])+lim)

drawTriangle(genTriangle(),'g')

```

```python
def rotTriangle(tr,angle):
    theta = (angle/180.) * numpy.pi
    rotMatrix = numpy.array([[numpy.cos(theta), -numpy.sin(theta)], 
                         [numpy.sin(theta),  numpy.cos(theta)]])
    a = np.dot(rotMatrix,tr[0])
    b = np.dot(rotMatrix,tr[1])
    c = np.dot(rotMatrix,tr[2])
    return np.array([a,b,c])

def cmRotTriangle(tr,angle):
    theta = (angle/180.) * numpy.pi
    rotMatrix = numpy.array([[numpy.cos(theta), -numpy.sin(theta)], 
                         [numpy.sin(theta),  numpy.cos(theta)]])
    cm = (tr[0]+tr[1]+tr[2])/3.
    a = np.dot(rotMatrix,tr[0]-cm)+cm
    b = np.dot(rotMatrix,tr[1]-cm)+cm
    c = np.dot(rotMatrix,tr[2]-cm)+cm
    return np.array([a,b,c])
```

```python
triforce = genTriangle()
drawTriangle(triforce,'b')
drawTriangle(rotTriangle(triforce,20),'g')
drawTriangle(cmRotTriangle(triforce,20),'r')
```

### Conclusion: Rotations about the origin and about the center of mass have the same effect on relative vectors $r_i-r_j$
